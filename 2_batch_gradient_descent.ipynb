```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Cost Function with Multiple Minima

## $$g(x) = x^4 - 4x^2 + 5$$

```python
def g(x):
    return x**4 - 4*x**2 + 5
```

# Derivative of the Cost Function

## $$g'(x) = 4x^3 - 8x$$

```python
def dg(x):
    return 4*x**3 - 8*x
```

```python
# Make data
x_1 = np.linspace(start=-2, stop=2, num=1000)
```

```python
# Plot cost function and derivative side by side

plt.figure(figsize=[15,5])

# 1 Chart: Cost Function
plt.subplot(1, 2, 1)

plt.xlim(-2, 2)
plt.ylim(0.5, 5.5)

plt.title('Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)

plt.plot(x_1, g(x_1), color='blue', linewidth=3)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.xlim(-2, 2)
plt.ylim(-6, 8)

plt.title('Slope of the Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()

plt.plot(x_1, dg(x_1), color='skyblue', linewidth=3)

plt.show()
```

# Batch Gradient Descent

```python
def gradient_descent(derivative_func, initial_guess, 
                     learning_rate=0.02, precision=0.001):
    
    new_x = initial_guess # starting point

    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(500):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - learning_rate * gradient

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        step_size = abs(new_x - previous_x)
        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

```python
local_min, x_list, deriv_list = gradient_descent(derivative_func=dg, initial_guess=0.5,
                                                 learning_rate=0.01, precision=0.0001)
print('Starting point: x =', 0.5)
print('Local min occurs at: x =', local_min)
print('Cost at this minimum is:', g(local_min))
print('Slope or df(x) value at this point is:', dg(local_min))
print('Number of steps:', len(x_list))
```

```python
local_min, x_list, deriv_list = gradient_descent(derivative_func=dg, initial_guess=-0.1)
print('Starting point: x =', -0.1)
print('Local min occurs at: x =', local_min)
print('Cost at this minimum is:', g(local_min))
print('Slope or df(x) value at this point is:', dg(local_min))
print('Number of steps:', len(x_list))
```

```python
# Calling gradient descent function with a positive starting point
local_min, x_list, deriv_list = gradient_descent(derivative_func=dg, 
                                                 initial_guess=0.1)

# Superimpose the gradient descent calculations on plot

plt.figure(figsize=[15,5])

# 1 Chart: Cost Function
plt.subplot(1, 2, 1)

plt.xlim(-2, 2)
plt.ylim(0.5, 5.5)

plt.title('Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)

plt.plot(x_1, g(x_1), color='blue', linewidth=3, alpha=0.8)
plt.scatter(x_list, g(np.array(x_list)), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.xlim(-2, 2)
plt.ylim(-6, 8)

plt.title('Slope of the Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()

plt.plot(x_1, dg(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, deriv_list, color='red', s=100, alpha=0.5)

plt.show()
print('Starting point: x =', 0.1)
print('Local min occurs at: x =', local_min)
print('Cost at this minimum is:', g(local_min))
print('Slope or df(x) value at this point is:', dg(local_min))
print('Number of steps:', len(x_list))
```

```python
# Calling gradient descent function with a negative starting point
local_min, x_list, deriv_list = gradient_descent(derivative_func=dg, 
                                                 initial_guess=-0.1)

# Superimpose the gradient descent calculations on plot

plt.figure(figsize=[15,5])

# 1 Chart: Cost Function
plt.subplot(1, 2, 1)

plt.xlim(-2, 2)
plt.ylim(0.5, 5.5)

plt.title('Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)', fontsize=16)

plt.plot(x_1, g(x_1), color='blue', linewidth=3, alpha=0.8)
plt.scatter(x_list, g(np.array(x_list)), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.xlim(-2, 2)
plt.ylim(-6, 8)

plt.title('Slope of the Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dg(x)', fontsize=16)
plt.grid()

plt.plot(x_1, dg(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, deriv_list, color='red', s=100, alpha=0.5)

plt.show()
print('Starting point: x =', -0.1)
print('Local min occurs at: x =', local_min)
print('Cost at this minimum is: ', g(local_min))
print('Slope or dg(x) value at this point is:', dg(local_min))
print('Number of steps: ', len(x_list))
```
